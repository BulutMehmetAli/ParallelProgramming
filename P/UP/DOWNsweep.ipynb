```python
def sequential_prefix_sum(arr):
    result = [0] * len(arr)
    result[0] = arr[0]
    
    for i in range(1, len(arr)):
        result[i] = result[i-1] + arr[i]
    
    return result

arr = [1, 2, 3, 4, 5, 6, 7, 8]
print(sequential_prefix_sum(arr))
```

```python
import math

def parallel_prefix_sum(arr):
    n = len(arr)
    x = arr.copy()
    
    # ===== UPSWEEP FAZI =====
    # Ağacı aşağıdan yukarıya doldur
    for d in range(int(math.log2(n))):
        step = 2 ** (d + 1)        # Her turda atlama miktarı
        half_step = 2 ** d         # Yarı adım
        
        # Bu işlemler gerçek hayatta PARALEL çalışır
        for k in range(0, n, step):
            x[k + step - 1] += x[k + half_step - 1]
        
        print(f"Upsweep Tur {d+1}: {x}")
    
    # ===== DOWNSWEEP FAZI =====
    # Kökü sıfırla
    x[n - 1] = 0
    
    # Ağacı yukarıdan aşağıya dağıt
    for d in range(int(math.log2(n)) - 1, -1, -1):
        step = 2 ** (d + 1)
        half_step = 2 ** d
        
        # Bu işlemler de gerçek hayatta PARALEL çalışır
        for k in range(0, n, step):
            temp = x[k + half_step - 1]
            x[k + half_step - 1] = x[k + step - 1]
            x[k + step - 1] += temp
        
        print(f"Downsweep Tur {d+1}: {x}")
    
    return x

arr = [1, 2, 3, 4, 5, 6, 7, 8]
print("Girdi:  ", arr)
print("---")
result = parallel_prefix_sum(arr)
print("---")
print("Çıktı:  ", result)
```
